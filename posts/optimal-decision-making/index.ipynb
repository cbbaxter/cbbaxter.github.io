```python
import pandas as pd
url = "https://raw.githubusercontent.com/PhilChodrow/ml-notes/main/data/credit-risk/train.csv"
df_train = pd.read_csv(url)
```

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

df_train.head()
```

```python
# Part B: Explore the Data

mean_income_by_home_ownership = df_train.groupby("loan_intent")["person_income"].mean()
mean_income_by_home_ownership.plot(kind = "bar")
plt.xlabel('Loan Intent')
plt.ylabel('Mean Income')
plt.show()
```

```python
plt.scatter(df_train["loan_amnt"], df_train["loan_percent_income"], c = df_train["loan_status"])
plt.xlabel('Loan Amount')
plt.ylabel('Loan Percent of Income')
plt.title('Scatter Plot of Loan Amount vs Loan Percent of Income')
plt.colorbar(label='Loan Status')

plt.show()
```
