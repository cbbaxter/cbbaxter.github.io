```python
# Part 1: Data Access
import pandas as pd
url = "https://gitlab.com/labsysmed/dissecting-bias/-/raw/master/data/data_new.csv?inline=false"
df = pd.read_csv(url)

df.head()
```

```python
# Part B: Reproduce Figure 1

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns


b_subset = df.loc[df["race"] == "black"]
b_percentiles = np.percentile(b_subset["risk_score_t"].values, np.arange(0, 100))

w_subset = df.loc[df["race"] == "white"]
w_percentiles = np.percentile(w_subset["risk_score_t"].values, np.arange(0, 100))

b_mean_c_i = []
w_mean_c_i = []

for i in range(0, 100):
    b_indices = np.where(b_subset["risk_score_t"].values >= i & b_subset["risk_score_t"].values < i+1)
    b_mean = np.mean(b_subset[b_indices].mean())
    b_mean_c_i.append(b_mean)

    w_indices = np.where(w_subset["risk_score_t"].values >= i & w_subset["risk_score_t"].values < i+1)
    w_mean = np.mean(w_subset[w_indices].mean())
    w_mean_c_i.append(w_mean)

plt.plot(X = b_percentiles, y = b_mean_c_i)
plt.show




```
